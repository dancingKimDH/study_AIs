```python
import pandas as pd
```

```python
df_BR = pd.read_csv('../../codes/datasets/BreastCancerWisconsinDataSet.csv')
```

#### 전처리

```python
df_BR.info()
```

```python
df_BR[:20]
```

```python
df_BR = df_BR[['diagnosis', 'concavity_mean', 'perimeter_worst']]
```

```python
def MB (diagnosis):
    if diagnosis == 'M':
        return 0
    else: return 1
```

```python
df_BR['diagnosis'] = df_BR['diagnosis'].apply(MB)
df_BR
```

```python
df_BR.isnull().sum()
```

```python
target_train = df_BR[['diagnosis']]
features_train = df_BR[['concavity_mean', 'perimeter_worst']]
target_train.shape, features_train.shape
```

#### 모델학습

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(features_train, target_train)
```

```python
model.coef_, model.intercept_
```

#### 예측

```python
df_BR[20:30]
```

```python
model.predict(features_train[20:30])
```

```python
model.predict_proba(features_train[20:30])
```

#### 평가

```python
target_predict = model.predict(features_train)
target_predict
```

```python
# 정확도 산출
from sklearn.metrics import accuracy_score
accuracy_score(target_train, target_predict)
```
